```python
from dataloader import load_dataloader
import os
import pickle
from pathlib import Path
%pylab inline
import matplotlib.pyplot as plt
home = str(Path.home())
```

```python
input_h, input_w = 128, 128
dataset = 'sample'
cv_pretrained = False
top_k = 0

dict_file = os.path.join(home, 'data', 'sample', f'data_dict_{top_k}.pkl')
with open(dict_file, 'rb') as file:
    data_dict = pickle.load(file)
word_to_idx = data_dict['word_to_idx']
idx_to_word = data_dict['idx_to_word']
answer_word_to_idx = data_dict['answer_word_to_idx']
answer_idx_to_word = data_dict['answer_idx_to_word']
question_type_to_idx = data_dict['question_type_to_idx']
idx_to_question_type = data_dict['idx_to_question_type']

dataloader = load_dataloader(os.path.join(home, 'data'), dataset, True, 2, data_config=[input_h, input_w, 8, cv_pretrained, top_k])

for img, q, a, q_type in dataloader:
    if not cv_pretrained:
        img = img.transpose(1, 2).transpose(2, 3)
        img = img.cpu().numpy()
        print(img[0].shape)
        imgplot = plt.imshow(img[0])
        plt.show()
    print([idx_to_word[idx] for idx in q[0][0].tolist()])
    print(answer_idx_to_word[a[0].item()])
    print(idx_to_question_type[q_type[0].item()])
    break
```
